```python
# Importar librerias
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
# Variable independiente
X_train = np.array([[1],[2],[3]])
```

```python
# Variable dependiente
y_train = np.array([[100],[150],[125]])
```

```python
# Grafica 
import matplotlib.pyplot as plt
plt.plot(X_train, y_train)
```

```python
# Crear modelo regresion lineal
modelo_regre = linear_model.LinearRegression()
```

```python
modelo_regre.fit(X_train, y_train)
```

```python
# Para revisar la pendiente
print(modelo_regre.coef_)
```

```python
# Para revisar la intersección de la línea
print(modelo_regre.intercept_)
```

```python
# vamos a generar predicciones
modelo_regre.predict([[4]])
```

```python
X = np.array([[1],[2],[3],[4]])
y = np.array([[100],[150],[125],[150]])
plt.plot(X, y)
```

```python
modelo_regre.predict([[5]])
```
